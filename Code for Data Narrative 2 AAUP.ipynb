Q1. Which state in America has the minimum 
number of schools? How can we visualise the 
relative number of schools in various states of 
America against the names of the states?

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy
df=pd.read_csv('http://lib.stat.cmu.edu/datasets/colleges/aaup.data')

state=numpy.array(df['AK'])
statenames, counts = numpy.unique(state, return_counts=True)
counts[0]+=1

plt.figure(figsize=(20,10))
plt.bar(statenames,counts)
plt.title("STATE v/s NUMBER OF UNIVERSITIES IN THAT STATE")
plt.xlabel("STATE")
plt.ylabel("NUMBER OF UNIVERSITIES")
plt.grid()
plt.show()

plt.figure(figsize=(20,20))
plt.pie(counts, labels=statenames)
plt.title("STATE v/s NUMBER OF UNIVERSITIES IN THAT STATE")
plt.xlabel("STATE")
plt.ylabel("NUMBER OF UNIVERSITIES")
plt.show()

max=0
ind=0
for i in range(len(counts)):
  if counts[i]>max:
    ind=i
print(statenames[ind])
```

Q2. What is the relative proportion of the number of 
universities against the number of colleges and 
institutions? How many schools fall under the 
various categorisations of group I, IIA, IIB, and 
VIIB? How can we visualise this data?

```python
##Number of universities, colleges, institutes

import pandas as pd
import matplotlib.pyplot as plt
import numpy
df=pd.read_csv('http://lib.stat.cmu.edu/datasets/colleges/aaup.data')

college_names=numpy.array(df['Alaska Pacific University'])
numpy.append(college_names,"Alaska Pacific University")

univ_counter=0
college_counter=0
institute_counter=0


for item in college_names:
  if "Univ" in item:
    univ_counter+=1
  elif "College" in item:
    college_counter+=1
  else:
    institute_counter+=1

counts=[institute_counter,college_counter,univ_counter]
types=["Institute","College",'University']


plt.figure()
plt.pie(counts, labels=types)
plt.title("The Relative Number of Various Types of Education Centres")
plt.show()

print("Number of Universities:",univ_counter)
print("Number of College:",college_counter)
print("Number of Institute:",institute_counter)
```

```python
#Type
import pandas as pd
import matplotlib.pyplot as plt
import numpy
df=pd.read_csv('http://lib.stat.cmu.edu/datasets/colleges/aaup.data')

total_type_roman_number=numpy.array(df['IIB'])
typenames, counts = numpy.unique(total_type_roman_number, return_counts=True)
counts[0]+=1

plt.figure()
plt.bar(typenames,counts)
plt.title("Various Types of Education Centres v/s Number of Education Centres of That Type")
plt.xlabel("Types of Education Centres")
plt.ylabel("Number of Education Centres")
plt.grid()
plt.show()

plt.figure()
plt.pie(counts, labels=typenames)
plt.title("The Relative Number of Various Types of Education Centres")
plt.show()
```

Q3. What is the mean, variance and standard 
deviation of the average salaries of full professors, 
associate professors and assistant professors? How 
can we visualise it with a histogram? How would a 
smooth plot of the probability distribution of the 
salary of the faculties against the probability of 
finding the faculty with that salary look like, and 
what can we analyse from that?

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy
df=pd.read_csv('http://lib.stat.cmu.edu/datasets/colleges/aaup.data')
for index, row in df.iterrows():
  if row["454"]=="*":
    df.drop(index)
  else:
    row["454"]=int(row["454"])
avg_salary_full=numpy.array(df['454'])
numpy.append(avg_salary_full,454)
avg_salary_full1=[]
for item in avg_salary_full:
  if item!="*":
    avg_salary_full1.append(item)
avg_salary_full1=list(map(int, avg_salary_full1))
avg_salary_full1.sort()
avg_salary_full1=numpy.array(avg_salary_full1)

print("Mean Salary of Full Professors:", numpy.mean(avg_salary_full1))
print("Variance Salary of Full Professors:", numpy.var(avg_salary_full1))
print("Standard deviation Salary of Full Professors:", numpy.std(avg_salary_full1))


plt.figure()
plt.hist(avg_salary_full1, bins=numpy.arange(200,1200,10))
plt.xlabel("Average Salary of Full Professors (in 100$ per year)")
plt.ylabel("Number of Professors")
plt.title("Histogram: Average Salary v/s No. of Professors")
plt.show()

plt.figure()
avg_salary_full1=pd.Series(avg_salary_full1)
avg_salary_full1.plot.kde()
plt.xlabel("Average Salary of Full Professors (in 100$ per year)")
plt.ylabel("Probability of Professors")
plt.title("Smooth Curve: Average Salary v/s Probability of Professors")
plt.show()

```

Q6. What is the mean, variance and standard 
deviation of the expenditure of the schools towards 
salary and compensation? What would a histogram 
of various schools in this aspect look like? What
would the probability distribution curve of getting a 
school having an expenditure of a particular value 
look like, and what can we analyse from that?


```python
##Total Expenditure of the schools towards salries and compensations
#Number of faculty
import pandas as pd
import matplotlib.pyplot as plt
import numpy
df=pd.read_csv('http://lib.stat.cmu.edu/datasets/colleges/aaup.data')

for x in df.index:
  if ((df.loc[x, "382.1"] == "*") or (df.loc[x, "487"] == "*") or (df.loc[x, "32"] == "*")):
    df.drop(x, inplace = True)

total_salary=[]

for x in df.index:
  total_salary.append((df.loc[x,"382.1"])*(df.loc[x,"32"])*100)

total_salary.sort()
total_salary=numpy.array(total_salary)
print("Mean Expenditure of Schools towards Salary (100$):", numpy.mean(total_salary))
print("Variance Expenditure of Schools towards Salary(100$):", numpy.var(total_salary))
print("Standard deviation Expenditure of Schools towards Salary(100$):", numpy.std(total_salary))

plt.figure()
plt.hist(total_salary)
plt.xlabel("Expenditure of Schools towards Salary (100$)")
plt.ylabel("Number of Schools")
plt.title("Histogram: Expenditure of Schools towards Salary v/s No. of Schools")
plt.show()

plt.figure()
avg_salary_full1=pd.Series(total_salary)
avg_salary_full1.plot.kde()
plt.xlabel("Expenditure of Schools towards Salary (100$)")
plt.ylabel("Probability of Schools")
plt.title("Smooth Curve: Expenditure of Schools towards Salary v/s No. of Schools")
plt.show()
```
