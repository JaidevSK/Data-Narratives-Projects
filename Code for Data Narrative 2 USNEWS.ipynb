Q1. How many schools in the survey are privately

the number of public schools? How can we visualise

```python
#Ratio of public to private

import pandas as pd
import matplotlib.pyplot as plt
import numpy
df=pd.read_csv('http://lib.stat.cmu.edu/datasets/colleges/usnews.data')

publicvsprivate=numpy.array(df['2'])
statenames, counts = numpy.unique(publicvsprivate, return_counts=True)
counts[0]+=1
print("Number of private schools=",counts[1])
print("Number of public schools=",counts[0])
print("ratio of the number of private and public schools=",counts[1]/counts[0])
types=["Public","Private"]
plt.figure()
plt.pie(counts, labels=types)
plt.title("Relative number of Public and Private Universities")
plt.show()
```

Q2. What is the mean, variance and standard 
deviation in the average SAT score in math and 
verbal abilities accepted by the schools? What 
would the visualisation of the average SAT score in 
math and verbal abilities look like against the 
number of schools at that score? What would the 
distribution of the probability of getting a school at 
a particular average SAT math and verbal score look 
like?

```python
##SAT score
import pandas as pd
import matplotlib.pyplot as plt
import numpy
df=pd.read_csv('http://lib.stat.cmu.edu/datasets/colleges/usnews.data')
for index, row in df.iterrows():
  if row["490"]=="*":
    df.drop(index)
  else:
    row["490"]=int(row["490"])
avg_salary_full=numpy.array(df['490'])
numpy.append(avg_salary_full,490)
avg_salary_full1=[]
for item in avg_salary_full:
  if item!="*":
    avg_salary_full1.append(item)
avg_salary_full1=list(map(int, avg_salary_full1))
avg_salary_full1.sort()
avg_salary_full1=numpy.array(avg_salary_full1)

print("Mean SAT Math score of all universities:", numpy.mean(avg_salary_full1))
print("Variance in the average SAT Math scores of all universities:", numpy.var(avg_salary_full1))
print("Standard deviation in the average SAT Math scores of all universities:", numpy.std(avg_salary_full1))



plt.figure()
plt.hist(avg_salary_full1, bins=numpy.arange(200,1200,10), histtype='step', color='red', label="Average Math Score")
plt.xlabel("Average SAT Math score")
plt.ylabel("Number of Universities at that Math score")
plt.title("Histogram: Average SAT score in Math v/s Number of universities being at that score")
plt.show()

plt.figure()
plt.hist(avg_salary_full1, bins=numpy.arange(200,1200,10), cumulative=True, histtype='step', color='red', label="Average Math Score")
plt.xlabel("Average SAT Math score")
plt.ylabel("Number of Universities that average Math score")
plt.title("Histogram: Average SAT score in Math v/s Number of universities accepting upto that average score")
plt.show()


plt.figure()
avg_salary_full1=pd.Series(avg_salary_full1)
avg_salary_full1.plot.kde()
plt.xlabel("Average SAT Math score")
plt.ylabel("Probability of a university being at that Math score")
plt.title("Smooth Curve: Average SAT score in Math v/s Probability of a university being at that score")
plt.show()

```

Q5. What is the mean, variance and standard 
deviation of the acceptance rate in various schools?
How can we visualise the trend of acceptance rate of 
various schools, and how would the probability 
distribution curve of getting a school having a 
particular rate of acceptance look like? What are the 
five schools having the least rate of acceptance? 
(Most difficult schools to get admission into.)


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy
df=pd.read_csv('http://lib.stat.cmu.edu/datasets/colleges/usnews.data')

for x in df.index:
  if ((df.loc[x, "193"] == "*") or (df.loc[x, "146"] == "*")):
    df.drop(x, inplace = True)

total_salary=[]
names=[]

for x in df.index:
  total_salary.append(int(df.loc[x,"146"])/int(df.loc[x,"193"]))
  names.append(df.loc[x,"Alaska Pacific University"])

selection_rate=total_salary.copy()

  
df1 = pd.DataFrame(selection_rate, index =names,columns =['rate'])
df1.sort_values(by=['rate'],ascending=True, inplace=True)
print(df1.head())

total_salary.sort()
total_salary=numpy.array(total_salary)
print("Mean Acceptance rate:", numpy.mean(total_salary))
print("Variance Acceptance rate:", numpy.var(total_salary))
print("Standard deviation Acceptance rate:", numpy.std(total_salary))

plt.figure()
plt.hist(total_salary)
plt.xlabel("Acceptance rate")
plt.ylabel("Number of Schools")
plt.title("Histogram: Acceptance rate v/s No. of Schools")
plt.show()

plt.figure()
avg_salary_full1=pd.Series(total_salary)
avg_salary_full1.plot.kde()
plt.xlabel("Acceptance rate")
plt.ylabel("Number of Schools")
plt.title("Smooth Curve: Acceptance rate v/s No. of Schools")
plt.show()
```
