```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('https://raw.githubusercontent.com/zygmuntz/goodbooks-10k/master/to_read.csv')

dicti={}
for index, row in df.iterrows():
  if row['book_id'] not in dicti.keys():
    dicti.update({row['book_id']:1})
  else:
    dicti[row['book_id']]+=1
bookid=dicti.keys()
no_of_users=dicti.values()
##probabilty of a person reading a book
total=sum(no_of_users)
total_list_of_number_of_readers=[(i/total) for i in no_of_users]
plt.figure()
plt.bar(bookid,total_list_of_number_of_readers)
plt.title("BOOK v/s PROBABILITY OF THAT BOOK BEING READ")
plt.xlabel("ID of the Book")
plt.ylabel("Probabilty of that book")
plt.show()
```

```python
##book vs avg rating
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv( 'https://raw.githubusercontent.com/zygmuntz/goodbooks-10k/master/ratings.csv' )

dicti={}

for index, row in df.iterrows():
  if row['book_id'] not in dicti.keys():
    dicti.update({row['book_id']:[]})
    dicti[row['book_id']].append(row['rating'])
  else:
    dicti[row['book_id']].append(row['rating'])
bookid=dicti.keys()
list_of_ratings=list(dicti.values())
for i in range(len(list_of_ratings)):
  list_of_ratings[i]=np.mean(np.array(list_of_ratings[i]))

plt.figure()
plt.bar(bookid,list_of_ratings)
plt.title("BOOK v/s AVERAGE RATING OF THAT BOOK")
plt.xlabel("ID of the Book")
plt.ylabel("Average Rating of that book")
plt.show()
```

```python
##language
import matplotlib.pyplot as plt
import pandas as pd
import numpy
df = pd.read_csv( 'https://raw.githubusercontent.com/zygmuntz/goodbooks-10k/master/books.csv' )


for x in df.index:
  if type(df.loc[x, "language_code"]) != str :
    df.drop(x, inplace = True)

names=numpy.array(df['language_code'])
values, counts = np.unique(names, return_counts=True)

plt.figure(figsize=(20,10))
plt.bar(values,counts)
plt.title("LANGUAGE v/s NUMBER OF BOOKS IN THAT LANGUAGE")
plt.xlabel("LANGUAGE")
plt.ylabel("NUMBER OF BOOKS")
plt.grid()
plt.show()
```

```python
# original_publication_year
import matplotlib.pyplot as plt
import pandas as pd
import numpy
df = pd.read_csv( 'https://raw.githubusercontent.com/zygmuntz/goodbooks-10k/master/books.csv' )



names=numpy.array(df['original_publication_year'])
values, counts = np.unique(names, return_counts=True)

plt.figure(figsize=(20,5))
plt.plot(values,counts)
plt.title("YEAR OF ORIGINAL PUBLICATION v/s NUMBER OF BOOKS IN THAT YEAR")
plt.xlabel("YEAR OF ORIGINAL PUBLICATION")
plt.ylabel("NUMBER OF BOOKS")
plt.grid()
plt.show()

df = pd.DataFrame({'Probability (KDE curve)':names})
ax = df.plot.kde()
plt.title("Probability of book being published in that year")
plt.xlabel("Year")
plt.ylabel("Probability")
plt.show()
```

```python
# probabilty original_publication_year
import matplotlib.pyplot as plt
import pandas as pd
import numpy
df = pd.read_csv( 'https://raw.githubusercontent.com/zygmuntz/goodbooks-10k/master/books.csv' )



names=numpy.array(df['original_publication_year'])
value, counts = np.unique(names, return_counts=True)
N=len(names)
counts=counts/N

plt.figure(figsize=(10,5))
plt.plot(value,counts)
plt.title("YEAR OF ORIGINAL PUBLICATION v/s PROBABILITY FOR A BOOK TO BE PUBLISHED IN THAT YEAR")
plt.xlabel("YEAR OF ORIGINAL PUBLICATION")
plt.ylabel("PROBABILTY")
plt.grid()
plt.show()




plt.figure()

cumulative = np.cumsum(abs(value))
  

plt.plot(value, cumulative, label="Cumulative Probability Distribution")
plt.title("YEAR OF ORIGINAL PUBLICATION v/s CUMULATIVE NUMBER OF BOOKS PUBLISHED BEFORE TILL YEAR")
plt.xlabel("YEAR OF ORIGINAL PUBLICATION")
plt.ylabel("Cumulative Number of Books")
plt.grid()
plt.legend()


plt.figure()
# print((cumulative))
#From this, we get the value of cumulative as 499043.
plt.plot(value, cumulative/499043, label="Cumulative Number of Books")
plt.title("YEAR OF ORIGINAL PUBLICATION v/s CUMULATIVE PROBABILITY OF A BOOK BEING PUBLISHED BEFORE THAT YEAR\nCDF")
plt.xlabel("YEAR OF ORIGINAL PUBLICATION")
plt.ylabel("Cumulative Probability")
plt.grid()
plt.legend()
```

```python
#BOOKS_COUNT
import matplotlib.pyplot as plt
import pandas as pd
import numpy
df = pd.read_csv( 'https://raw.githubusercontent.com/zygmuntz/goodbooks-10k/master/books.csv' )

BOOKS_COUNT=df['books_count']
BOOK_ID=df['book_id']
TOTAL_BOOKS=sum(BOOKS_COUNT)

plt.figure()
plt.plot(BOOK_ID, BOOKS_COUNT, label="Number of Copies of Each book")
plt.title("BOOK ID v/s NUMBER OF COPIES OF THAT BOOK")
plt.xlabel("ID NUMBER OF THE BOOK")
plt.ylabel("NUMBER OF COPIES OF THAT BOOK")
plt.grid()
plt.legend()

#PROBABILITY OF GETTING A BOOK AT RANDOM FROM THE COLLECTION OF BOOKS
plt.figure()
plt.plot(BOOK_ID, BOOKS_COUNT/TOTAL_BOOKS, label="Probability of getting a Particular Book")
plt.title("BOOK ID v/s PROBABILITY OF GETTING THAT BOOK")
plt.xlabel("ID NUMBER OF THE BOOK")
plt.ylabel("PROBABILITY OF GETTING THAT BOOK")
plt.grid()
plt.legend()


#CUMULATIVE PROBABILITY OF GETTING A BOOK AT RANDOM HAVING ID LESS THAN A PARTICULAR NUMBER FROM THE COLLECTION OF BOOKS
plt.figure()

cumulative = np.cumsum(abs(BOOKS_COUNT/TOTAL_BOOKS))
plt.plot(BOOK_ID, cumulative, label="CDF")
plt.title("BOOK ID v/s CUMULATIVE PROBABILITY OF GETTING A BOOK LESS THAN A PARTICULAR ID")
plt.xlabel("ID NUMBER OF THE BOOK")
plt.ylabel("CUMULATIVE PROBABILITY")
plt.grid()
plt.legend()
```
